```python
import pandas as pd
```

```python
df = pd.read_csv('ownership.csv', delimiter = ';', decimal = ',')
cols = ['id. Response ID', 'submitdate. Date submitted','Carmain',
       'Previous PHEVs', 'Previous BEVs', 'Previous HEVs', 'Previous CNGs',
       'Household Income',
       'Importance of reducing greenhouse gas emissions (-3 not important, 3 important)',
       'Home ownership (own 1)', 'Home Type (detached 1)',
       'Highest Level of Education', 'Longest trip in the last 12 months',
       'Number of trips over 200 miles in the last 12 months',
       'One-way commute distance', 'Number of people in the household', 'Age',
       'Gender (Male 1)', 'Number of vehicles in the household',
       'Annual VMT Estimate ']

df= (
    df
    [cols]
)
## column descriptions: Gender: Male=1, Female =0, N/A = -1
```

```python
cf = (
    df
    .assign(date = pd.to_datetime(df['submitdate. Date submitted'].fillna(method = 'ffill')))
    .assign(gender = df['Gender (Male 1)'].fillna(-1).astype('int16'))
    .assign(car = df['Carmain'].fillna('N/A'))
    .assign(prev_green_vehicle = df['Previous PHEVs'].fillna(0)+ \
                                 df['Previous BEVs'].fillna(0) + \
                                 df['Previous HEVs'].fillna(0) + \
                                 df['Previous CNGs'].fillna(0))
    .drop(columns = ['submitdate. Date submitted','Gender (Male 1)', 'Carmain','Previous PHEVs', \
         'Previous BEVs', 'Previous HEVs', 'Previous CNGs'])

)
```

```python
### basic statistics on dataset
### 1) What is the average income?
avg_inc = cf['Household Income'].dropna().mean()
print(f'The average annual income of an electric car owner is {avg_inc:.3f}$')

### 2) Percentage Male vs Female 
perc_male = 100*cf.gender[cf.gender != -1].mean()
print(f'{perc_male:.2f}\% of electric car owners are men.')

### 3) Age
avg_age = cf['Age'].dropna().mean()
print(f'The average electric car owner is {avg_age:.2f} years old.')
```
